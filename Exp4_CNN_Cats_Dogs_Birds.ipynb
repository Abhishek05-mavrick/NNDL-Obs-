## Code

```python
# Experiment 4 – CNN: Cats, Dogs, Birds
# Abhishek V Ammanagi | 2023BCSE07AED500

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, Model
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

print("=" * 55)
print("  Abhishek V Ammanagi  |  2023BCSE07AED500")
print("  Experiment 4: CNN – Cats, Dogs, Birds")
print("=" * 55)

IMG_SIZE    = 64
NUM_CLASSES = 3
EPOCHS      = 15
CLASS_NAMES = ['Birds','Cats','Dogs']
np.random.seed(42)

# Load CIFAR-10 dataset - classes 2(bird), 3(cat), 5(dog)
from tensorflow.keras.datasets import cifar10
(X_train_all, y_train_all), (X_test_all, y_test_all) = cifar10.load_data()
y_train_all = y_train_all.flatten()
y_test_all = y_test_all.flatten()
# Filter for bird(2), cat(3), dog(5)
class_indices = [2, 3, 5]
mask_train = np.isin(y_train_all, class_indices)
mask_test = np.isin(y_test_all, class_indices)
X_tr_filtered = X_train_all[mask_train].astype('float32') / 255.0
y_tr_filtered = y_train_all[mask_train]
X_te_filtered = X_test_all[mask_test].astype('float32') / 255.0
y_te_filtered = y_test_all[mask_test]
# Remap labels to 0,1,2
y_tr_remapped = np.array([class_indices.index(c) for c in y_tr_filtered])
y_te_remapped = np.array([class_indices.index(c) for c in y_te_filtered])
# Resize to 64×64 and split
from tensorflow.image import resize
X_tr_resized = resize(X_tr_filtered, (IMG_SIZE, IMG_SIZE)).numpy()
X_te_resized = resize(X_te_filtered, (IMG_SIZE, IMG_SIZE)).numpy()
split_idx = int(0.8 * len(X_tr_resized))
X_train = X_tr_resized[:split_idx]
y_train = y_tr_remapped[:split_idx]
X_val = X_tr_resized[split_idx:]
y_val = y_tr_remapped[split_idx:]
y_tr_c  = tf.keras.utils.to_categorical(y_train, NUM_CLASSES)
y_va_c  = tf.keras.utils.to_categorical(y_val,   NUM_CLASSES)

model = models.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',padding='same',
                  input_shape=(IMG_SIZE,IMG_SIZE,3)),
    layers.Conv2D(32,(3,3),activation='relu',padding='same'),
    layers.MaxPooling2D(2), layers.Dropout(0.25),
    layers.Conv2D(64,(3,3),activation='relu',padding='same'),
    layers.Conv2D(64,(3,3),activation='relu',padding='same'),
    layers.MaxPooling2D(2), layers.Dropout(0.25),
    layers.Conv2D(128,(3,3),activation='relu',padding='same'),
    layers.MaxPooling2D(2), layers.Dropout(0.4),
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(NUM_CLASSES, activation='softmax')
])
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()

history = model.fit(X_train, y_tr_c,
                    validation_data=(X_val, y_va_c),
                    epochs=EPOCHS, batch_size=32)

loss, acc = model.evaluate(X_val, y_va_c, verbose=0)
y_pred    = np.argmax(model.predict(X_val), axis=1)
print(f"\nValidation Accuracy : {acc:.4f}")
print(classification_report(y_val, y_pred, target_names=CLASS_NAMES))

fig, ax = plt.subplots(1,2,figsize=(12,4))
ax[0].plot(history.history['accuracy'],     label='Train')
ax[0].plot(history.history['val_accuracy'], label='Val')
ax[0].set_title('Accuracy'); ax[0].legend(); ax[0].grid(True)
ax[1].plot(history.history['loss'],     label='Train')
ax[1].plot(history.history['val_loss'], label='Val')
ax[1].set_title('Loss'); ax[1].legend(); ax[1].grid(True)
plt.suptitle("Abhishek V Ammanagi | 2023BCSE07AED500", fontsize=9)
plt.tight_layout(); plt.savefig('exp4_curves.png',dpi=300); plt.show()

# Feature maps – first conv layer
fm_model = Model(inputs=model.inputs, outputs=model.layers[0].output)
fmaps    = fm_model.predict(X_val[0:1])
fig, axes = plt.subplots(4,8,figsize=(16,8))
for i, ax2 in enumerate(axes.flat):
    if i < fmaps.shape[-1]:
        ax2.imshow(fmaps[0,:,:,i], cmap='viridis')
    ax2.axis('off')
plt.suptitle("First Conv Layer Feature Maps\nAbhishek V Ammanagi | 2023BCSE07AED500", fontsize=9)
plt.tight_layout(); plt.savefig('exp4_fmaps.png',dpi=300); plt.show()

cm   = confusion_matrix(y_val, y_pred)
disp = ConfusionMatrixDisplay(cm, display_labels=CLASS_NAMES)
disp.plot(cmap='Blues')
plt.title("Confusion Matrix – Cats/Dogs/Birds")
plt.savefig('exp4_cm.png',dpi=300); plt.show()
print("Abhishek V Ammanagi | 2023BCSE07AED500")
```

```python
# ===== OUTPUT VISUALIZATION =====
print("\n" + "="*60)
print("DETAILED RESULTS - CNN CATS/DOGS/BIRDS")
print("="*60)

# Display sample predictions with confidence (deterministic ordering)
num_samples = min(12, len(X_val))
fig, axes = plt.subplots(3, 4, figsize=(16, 12))

for idx, ax in enumerate(axes.flat):
    if idx >= num_samples:
        ax.axis('off')
        continue
        
    sample_idx = idx
    img = X_val[sample_idx]
    ax.imshow(img)
    
    true_label = y_val[sample_idx]
    pred_label = y_pred[sample_idx]
    pred_probs = model.predict(X_val[sample_idx:sample_idx+1], verbose=0)[0]
    confidence = np.max(pred_probs)
    
    color = 'green' if true_label == pred_label else 'red'
    ax.set_title(f'True: {CLASS_NAMES[true_label]} | Pred: {CLASS_NAMES[pred_label]}\nConf: {confidence:.2%}', 
                 color=color, fontweight='bold', fontsize=9)
    ax.axis('off')

plt.suptitle("Sample Predictions - CIFAR-10 Animal Classification\nAbhishek V Ammanagi | 2023BCSE07AED500", 
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('exp4_sample_predictions.png', dpi=300, bbox_inches='tight')
plt.show()

# Accuracy by class
print("\n📊 ACCURACY BY CLASS:")
for i in range(NUM_CLASSES):
    class_mask = y_val == i
    if class_mask.sum() > 0:
        class_acc = (y_pred[class_mask] == y_val[class_mask]).mean()
        print(f"  {CLASS_NAMES[i]}: {class_acc:.1%} ({class_mask.sum()} samples)")

# Per-class statistics
print("\n📈 OVERALL METRICS:")
correct = (y_pred == y_val).sum()
total = len(y_val)
print(f"  ✓ Total Correct: {correct}/{total}")
print(f"  ✓ Validation Accuracy: {acc:.2%}")
print(f"  ✓ Model Parameters: {model.count_params():,}")

# Display class samples
print("\n" + "="*60)
print("SAMPLE IMAGES BY CLASS")
print("="*60)
fig, axes = plt.subplots(NUM_CLASSES, 6, figsize=(16, 8))

for class_idx in range(NUM_CLASSES):
    class_indices = np.where(y_val == class_idx)[0]
    sample_count = min(6, len(class_indices))
    
    for i in range(6):
        ax = axes[class_idx, i]
        if i < sample_count:
            ax.imshow(X_val[class_indices[i]])
            ax.set_title(CLASS_NAMES[class_idx], fontsize=9)
        ax.axis('off')

plt.suptitle("Sample Animals from CIFAR-10\nAbhishek V Ammanagi | 2023BCSE07AED500", 
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('exp4_class_samples.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✅ CNN Animal Classification Complete!")
print("   Output files: exp4_curves.png, exp4_fmaps.png, exp4_cm.png, exp4_sample_predictions.png, exp4_class_samples.png")
```
